```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
import seaborn as sns
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2020,1,1)
```

```python
end = datetime.datetime(2023,1,1)
```

```python
BAC = yf.download('BAC',start,end)
```

```python
BAC
```

```python
C = yf.download('C',start,end)
GS = yf.download('GS',start,end)
JPM = yf.download('JPM',start,end)
```

```python
tickers = ['BAC','C','GS','JPM']
```

```python
bank_stock = pd.concat([BAC,C,GS,JPM],axis=1,keys=tickers)
```

```python
bank_stock
```

```python
bank_stock.columns.names = ['bank_ticker','stock_info']
```

```python
bank_stock
```

```python
bank_stock.xs(key='Close',axis=1,level='stock_info').max()
```

```python
bank_stock.xs(key='Close',axis=1,level='stock_info').min()
```

```python
bank_stock.xs(key='Close',axis=1,level='stock_info').idxmax()
```

```python
bank_stock.xs(key='Close',axis=1,level='stock_info').idxmin()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + 'return'] = bank_stock[tick]['Close'].pct_change()
```
